# Merging all JSON Files Collected

```python
import pandas as pd
import numpy as np
import json
import os

# --------- Panda Settings ------------ #
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_colwidth', -1)
```

```python
DATA_DIR = '../data'
BASE_MEMES = 'memefly_data.json'
BLEACHED_DATA = 'bleached_data.json'

DATA_BASE_PATH = os.path.join(DATA_DIR, BASE_MEMES)
DATA_BLEACH_PATH = os.path.join(DATA_DIR, BLEACHED_DATA)


print(DATA_BASE_PATH)
print(DATA_BLEACH_PATH)
```

### Load base meme names

```python
base_memes = pd.read_json(DATA_BASE_PATH, orient='records', typ='frame')
base_memes.head()
```

```python
base_memes.info()
```

### Load bleach data

```python
bleach_memes = pd.read_json(DATA_BLEACH_PATH, orient='records', typ='frame')
bleach_memes.head()
```

```python
# Remove any straggling punctuation
import string
bleach_memes['base_meme_name'] = bleach_memes['base_meme_name'].str.translate(str.maketrans('', '', string.punctuation))
bleach_memes.sample(15)
```

```python
# Edit base_meme_name column
# add - where there are spaces
bleach_memes['base_meme_name'] = bleach_memes['base_meme_name'].str.strip().str.lower().str.replace(' ', '-')
bleach_memes.sample(15)
```

```python
# How many memes in each dataframe?
base_memes['meme_name'].value_counts().sum(), bleach_memes['base_meme_name'].value_counts().sum()
```

```python
# Merge merged & bleach_memes
# left on meme_name, right on base_meme_name
merged = base_memes.merge(bleach_memes, 
                      left_on='meme_name', 
                      right_on='base_meme_name',
                      )
merged = merged.drop(columns='base_meme_name')
merged = merged.rename(columns={'base_meme_text':'meme_text'})
print(f'Shape of DF: {merged.shape}')
print(merged.isnull().sum())
merged.head(15)
```

```python
grouped = merged.groupby('meme_name')
meme_txt_grouped = grouped['meme_text'].apply(list).to_frame().reset_index()
meme_txt_grouped.head(1)
```

```python
merged_de_dups = merged.drop_duplicates(subset='meme_name', keep='first').drop(columns='meme_text')
merged_de_dups.head()
```

```python
new_merged = merged_de_dups.merge(meme_txt_grouped, on='meme_name')
new_merged.head()
```

```python
new_merged.info()
```

```python
new_merged = new_merged.reset_index()
new_merged = new_merged.rename(columns={'index':'meme_id'})
new_merged.head()
```

```python
full_path = os.path.join(DATA_DIR, 'combined_data.json')
new_merged.to_json(full_path, orient='records')
```
