```python
import boto3
import json
from decouple import config
import os
import requests
import logging
import time
```

## Load `base_memes.json` file

```python
DATA_DIR = '../data'
BASE_MEME_IMGFLIP_LINKS = 'base_memes_from_imgflip.json'

with open(os.path.join(DATA_DIR, BASE_MEME_IMGFLIP_LINKS), 'r') as robj:
    base_memes = json.load(robj)
base_memes
```

## Dump to S3 `memefly-datalake` `raw-images` folder

```python
BASE_MEMES_S3_LINKS = 'base_memes_s3_links.json'
BUCKET_NAME = 'memefly'
S3_IMAGE_FOLDER = 'base-meme-images'   
REGION = 'us-east-2'
S3_BASE_URL = f'https://{BUCKET_NAME}.s3.{REGION}.amazonaws.com'
print(f'''
BUCKET_NAME = {BUCKET_NAME}
S3_IMAGE_FOLDER = {S3_IMAGE_FOLDER}
REGION  = {REGION }
S3_BASE_URL = {S3_BASE_URL}
''')
```

```python
def dump_imgflip_basememes_to_s3():
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Cafari/537.36'} 

    s3 = boto3.client(
        's3',
        aws_access_key_id=config('AWS_ACCESS_KEY'),
        aws_secret_access_key=config('AWS_SECRET_ACCESS_KEY')
    )
    
    for k, v in base_memes.items():
        print(f'Requesting base meme pic: {v}')
        pic = requests.get(v, headers=headers)
        pic_filename = v.split('/')[-1]
        pic_filename = pic_filename.lower()
        folder_path = '/'.join([S3_IMAGE_FOLDER, pic_filename])
        print(f'Saving to S3 folder path: {folder_path}')
        s3.put_object(Body = pic.content, Bucket = BUCKET_NAME, Key = folder_path)

dump_imgflip_basememes_to_s3()
```

```python
def save_base_memes_s3_links():
    s3 = boto3.client(
            's3',
            aws_access_key_id=config('AWS_ACCESS_KEY'),
            aws_secret_access_key=config('AWS_SECRET_ACCESS_KEY')
        )
    objs = s3.list_objects(Bucket=BUCKET_NAME)
    print(f'[INFO] Fetching S3 objects from {BUCKET_NAME}')
    data = {}
    for obj in objs['Contents']:
            if obj['Key'].endswith('.jpg'):
                s3_file_path = obj['Key']
                s3_link = S3_BASE_URL + '/' + s3_file_path
                s3_pic_name = s3_file_path.split('/')[1]
                data[s3_pic_name] = s3_link
    full_path = os.path.join(DATA_DIR, BASE_MEMES_S3_LINKS)
    print(f'[INFO] Saving file at {full_path}')
    with open(full_path, "w") as wobj:
        json.dump(data, wobj, indent=4)    

save_base_memes_s3_links()
```

```python
def download_base_memes_from_s3_bucket(pic_dir_path):
    
    if not os.path.exists(pic_dir_path):
        os.makedirs(pic_dir_path, exist_ok=True)
    
    bucket_name = 'memefly-datalake'
    s3_folder = 'base-meme-images'   
    
    s3 = boto3.client(
        's3',
        aws_access_key_id=config('AWS_ACCESS_KEY'),
        aws_secret_access_key=config('AWS_SECRET_ACCESS_KEY')
    )
    
    objs = s3.list_objects(Bucket=bucket_name)
    for obj in objs['Contents']:
        if obj['Key'].endswith('.jpg'):
            print(obj['Key'])
            s3_file_path = obj['Key']
            s3_folder = s3_file_path.split('/')[0]
            s3_pic_name = s3_file_path.split('/')[1]
            pic_download_loc = os.path.join(pic_dir_path, s3_pic_name)
            s3.download_file(bucket_name, s3_file_path, pic_download_loc)
            print(f'Downloaded to: {pic_download_loc}')


#download_base_memes_from_s3_bucket('../data/images')
```
